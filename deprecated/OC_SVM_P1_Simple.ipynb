```python
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import OneClassSVM
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, recall_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load preprocessed data
def load_preprocessed_file(file_path):
    """
    Load a preprocessed dataset file saved as a pickle (.pkl) file.

    Args:
        file_path (str): Path to the preprocessed file.

    Returns:
        dict: A dictionary containing 'features', 'labels', and 'file_paths'.
    """
    with open(file_path, "rb") as f:
        data = pickle.load(f)
    return data

# Paths to preprocessed datasets
TRAIN_FILE = "preprocessed_simple_data/train_100percent_features_labels.pkl"  # Update this path
EVAL_FILE = "preprocessed_simple_data/eval_100percent_features_labels.pkl"  # Update this path

# Load datasets
print("Loading preprocessed training data...")
train_data = load_preprocessed_file(TRAIN_FILE)
print(f"Training data loaded: {len(train_data['features'])} samples")

print("Loading preprocessed validation data...")
eval_data = load_preprocessed_file(EVAL_FILE)
print(f"Validation data loaded: {len(eval_data['features'])} samples")

# Extract features and labels
X_train = np.array(train_data["features"])
y_train = np.array(train_data["labels"])

X_eval = np.array(eval_data["features"])
y_eval = np.array(eval_data["labels"])

# Train One-Class SVM on training data (bona fide only)
print("Training One-Class SVM...")
X_train_bona_fide = X_train[y_train == 1]  # Filter only bona fide samples
# ocsvm = OneClassSVM(kernel='rbf', gamma='scale', nu=0.5) # ----------------------- 1
ocsvm = OneClassSVM(kernel='rbf', gamma=0.006, nu=0.2)
ocsvm.fit(X_train_bona_fide)

# Evaluate on the validation set
print("Evaluating on validation data...")
eval_scores = ocsvm.decision_function(X_eval)
eval_predictions = ocsvm.predict(X_eval)

# Map predictions to binary labels (1 = bona fide, 0 = spoof)
predicted_labels_eval = (eval_predictions == 1).astype(int)

# Compute metrics
roc_auc_eval = roc_auc_score(y_eval, eval_scores)
accuracy_eval = accuracy_score(y_eval, predicted_labels_eval)
recall_eval = recall_score(y_eval, predicted_labels_eval, pos_label=1)
conf_matrix = confusion_matrix(y_eval, predicted_labels_eval)
class_report = classification_report(y_eval, predicted_labels_eval, target_names=["Spoof", "Bona fide"])

# Print metrics
print(f"Validation ROC AUC Score: {roc_auc_eval:.4f}")
print(f"Validation Accuracy: {accuracy_eval:.4f}")
print(f"Validation Recall: {recall_eval:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Spoof", "Bona fide"], yticklabels=["Spoof", "Bona fide"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_eval, eval_scores)
plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_eval:.4f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for One-Class SVM on Validation Data")
plt.legend()
plt.grid()
plt.show()

# Visualize examples of predictions
num_examples = 5
print(f"Showing {num_examples} examples of validation predictions:")
for i in range(min(num_examples, len(X_eval))):
    print(f"Example {i + 1}: Predicted = {'Bona fide' if predicted_labels_eval[i] == 1 else 'Spoof'}, Actual = {'Bona fide' if y_eval[i] == 1 else 'Spoof'}")
    plt.figure(figsize=(10, 4))
    plt.plot(X_eval[i], marker="o")
    plt.title(f"Example {i + 1}: {'Correct' if predicted_labels_eval[i] == y_eval[i] else 'Incorrect'} Prediction")
    plt.xlabel("Feature Index")
    plt.ylabel("Feature Value")
    plt.grid(True)
    plt.show()

# Save the trained model
with open("trained_ocsvm_model.pkl", "wb") as model_file:
    pickle.dump(ocsvm, model_file)
print("Trained model saved successfully.")
```
