# Explanation of Preprocessing Steps

"""
### Dataset Preprocessing Overview

This notebook processes the ASVspoof 2019 Logical Access dataset to prepare it for machine learning models.
The preprocessing includes the following steps:

1. **Data Loading**:
   - The dataset is organized using protocol files that map each audio file to its label (`bonafide` or `spoof`).
   - These labels are critical for supervised learning tasks and are loaded into a dictionary for reference.

2. **Feature Extraction**:
   - We extract **MFCC (Mel-Frequency Cepstral Coefficients)** features from each audio file.
   - MFCC features provide a compact representation of the spectral properties of audio, commonly used in speech processing tasks.

3. **Shuffling and Dataset Splitting**:
   - After feature extraction, all data points (features, labels, and file paths) are shuffled to ensure that subsets do not have unintended biases.
   - This is especially important for creating smaller subsets (e.g., 10%, 50%) that are representative of the full dataset.

4. **Subsets Creation**:
   - Three subsets are created for each dataset (`train`, `dev`, `eval`):
     - **10% subset**: Useful for quick experimentation or debugging.
     - **50% subset**: Suitable for intermediate testing to balance speed and performance evaluation.
     - **100% subset**: Full dataset for final model training and evaluation.
   - These subsets allow flexibility in experimentation and reduce computational overhead during early-stage testing.

5. **Data Storage**:
   - The processed data (features, labels, and file paths) are saved in `.pkl` files for each subset.
   - This ensures that future notebooks can directly load the preprocessed data without reprocessing the audio files, saving significant computation time.

### Why This Approach?
- **Efficiency**: By storing preprocessed features, subsequent experiments can skip the computationally expensive audio processing step.
- **Flexibility**: Different subsets (10%, 50%, 100%) allow for tailored experimentation based on the current stage of model development.
- **Robustness**: Shuffling the data ensures subsets are representative of the overall dataset, reducing the risk of training/testing bias.

This preprocessing pipeline is designed to maximize the usability and reproducibility of the ASVspoof 2019 dataset for various machine learning models.
"""


```python
# Preprocessing script for the ASVspoof 2019 Logical Access Dataset

import os
import librosa
import numpy as np
import pandas as pd
import pickle
from tqdm import tqdm  # For progress bar
import random  # For shuffling

# Define paths
DATASET_DIR = "/mnt/c/DF/LA/"  # Update with the actual path
OUTPUT_DIR = "preprocessed_simple_data/"  # Directory to save intermediate files

# Ensure output directory exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Protocol files
TRAIN_PROTOCOL = os.path.join(DATASET_DIR, "ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.train.trn.txt")
DEV_PROTOCOL = os.path.join(DATASET_DIR, "ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.dev.trl.txt")
EVAL_PROTOCOL = os.path.join(DATASET_DIR, "ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.eval.trl.txt")

# Helper function to load protocol

def load_protocol(protocol_path):
    """
    Load the protocol file and return a DataFrame with relevant information.
    """
    data = pd.read_csv(
        protocol_path, 
        sep=" ", 
        header=None, 
        names=["speaker_id", "file_id", "system_id", "audio_type", "label"]
    )
    return data

# Feature extraction function
def extract_features(file_path, n_mfcc=20):
    """
    Extract MFCC features from the audio file.
    """
    audio, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    mfccs_mean = np.mean(mfccs.T, axis=0)  # Use mean of MFCCs
    return mfccs_mean

# Process dataset
def process_dataset(protocol, dataset_type):
    """
    Process the dataset and save the features and labels.
    """
    features = []
    labels = []
    file_paths = []

    for idx, row in tqdm(protocol.iterrows(), total=len(protocol), desc=f"Processing {dataset_type} dataset"):
        file_path = os.path.join(DATASET_DIR, f"ASVspoof2019_LA_{dataset_type}/flac", row['file_id'] + ".flac")
        if os.path.exists(file_path):
            mfcc_features = extract_features(file_path)
            features.append(mfcc_features)
            labels.append(1 if row['label'] == "bonafide" else 0)
            file_paths.append(file_path)
        else:
            print(f"File not found: {file_path}")

    # Combine and shuffle data
    combined = list(zip(features, labels, file_paths))
    random.shuffle(combined)
    features, labels, file_paths = zip(*combined)

    # Save the features and labels for different dataset sizes
    for fraction, fraction_name in [(0.1, "10"), (0.5, "50"), (1.0, "100")]:
        subset_size = int(len(features) * fraction)
        output_file = os.path.join(OUTPUT_DIR, f"{dataset_type}_{fraction_name}percent_features_labels.pkl")
        with open(output_file, "wb") as f:
            pickle.dump({
                "features": features[:subset_size], 
                "labels": labels[:subset_size], 
                "file_paths": file_paths[:subset_size]
            }, f)

        print(f"Processed and saved {fraction_name}% of {dataset_type} dataset to {output_file}")

# Load protocols
train_protocol = load_protocol(TRAIN_PROTOCOL)
dev_protocol = load_protocol(DEV_PROTOCOL)
eval_protocol = load_protocol(EVAL_PROTOCOL)

# Process and save datasets
process_dataset(train_protocol, "train")
process_dataset(dev_protocol, "dev")
process_dataset(eval_protocol, "eval")
```

# Import and visualization

```python
# Visualization of Preprocessed Data
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Define paths to the preprocessed data
PREPROCESSED_DIR = "preprocessed_simple_data/"  # Update with actual path

# Function to load preprocessed data
def load_preprocessed_data(file_path):
    """
    Load the preprocessed data from a pickle file.
    """
    with open(file_path, "rb") as f:
        data = pickle.load(f)
    return data

# Visualize a subset of the data
def visualize_data(preprocessed_file, num_samples=5):
    """
    Visualize features and labels from the preprocessed data.

    Args:
        preprocessed_file (str): Path to the preprocessed pickle file.
        num_samples (int): Number of samples to visualize.
    """
    # Load data
    data = load_preprocessed_data(preprocessed_file)
    features = data["features"]
    labels = data["labels"]

    # Select samples to visualize
    for i in range(min(num_samples, len(features))):
        feature = features[i]
        label = labels[i]

        # Plot the feature (assumes 1D MFCC vector)
        plt.figure(figsize=(10, 4))
        plt.plot(feature, marker="o")
        plt.title(f"Sample {i + 1}: Label {'Bonafide' if label == 1 else 'Spoof'}")
        plt.xlabel("Feature Index")
        plt.ylabel("Value")
        plt.grid(True)
        plt.show()

# Example usage
# Update with the actual file path to a preprocessed dataset file (e.g., train_10percent_features_labels.pkl)
PREPROCESSED_FILE = os.path.join(PREPROCESSED_DIR, "train_10percent_features_labels.pkl")
visualize_data(PREPROCESSED_FILE, num_samples=5)
```
